```python
def convertir_valor(valor):
    if isinstance(valor, float):
        return valor
    elif 'M' in valor:
        return float(valor.replace('$', '').replace('M', '')) * 1e6
    elif 'B' in valor:
        return float(valor.replace('$', '').replace('B', '')) * 1e9
    else:
        return float(valor.replace('$', ''))


df['Ganancias'] = df['Ganancias'].apply(convertir_valor)
df['Ganancias'].min()
```
